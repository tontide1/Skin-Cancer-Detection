```python
import pandas as pd
```

```python
df = pd.read_csv('data/raw/ISIC_2019_Test_GroundTruth.csv')
print(f"Số lượng ảnh ban đầu: {len(df)}")
df
```

```python
df.drop(['score_weight', 'validation_weight'], axis=1 ,inplace=True)
```

```python
df
```

```python
import os
from pathlib import Path
```

```python
IMAGE_FOLDER = Path('data/raw/ISIC_2019_Test_Input')

# Xóa các file ảnh có UNK = 1
unk_images = df[df['UNK'] == 1]['image'].tolist()
print(f"Số lượng ảnh UNK cần xóa: {len(unk_images)}")

deleted_count = 0
not_found_count = 0

for image_name in unk_images:
    for ext in ['.jpg', '.jpeg', '.png', '.JPG', '.JPEG', '.PNG']:
        image_path = IMAGE_FOLDER / f"{image_name}{ext}"
        if image_path.exists():
            os.remove(image_path)
            deleted_count += 1
            print(f"Đã xóa: {image_path}")
            break
    else:
        not_found_count += 1
        print(f"Không tìm thấy: {image_name}")

print(f"\nTổng kết:")
print(f"  - Đã xóa: {deleted_count} ảnh")
print(f"  - Không tìm thấy: {not_found_count} ảnh")
```

```python
# Lọc bỏ các hàng có UNK = 1 (giữ lại UNK = 0)
df_filtered = df[df['UNK'] == 0].copy()
print(f"Số lượng ảnh sau khi lọc UNK: {len(df_filtered)}")
```

```python
# Lưu file metadata đã xử lý
output_path = 'data/raw/ISIC_2019_Test_Metadata_Clean.csv'
os.makedirs(os.path.dirname(output_path), exist_ok=True)
df_filtered.to_csv(output_path, index=False)
print(f"Đã lưu file metadata tại: {output_path}")
print(f"Tổng số mẫu: {len(df_filtered)}")
```

```python
import pandas as pd
```

```python
df = pd.read_csv('data/raw/ISIC_2019_Test_Metadata_Clean.csv')
df
```

```python
df.drop(['UNK'], axis=1, inplace=True)
```

```python
DATA_DIR = 'data/processed/cropped_lesions_testset/'
```

```python
import os
classes = ['MEL', 'NV', 'BCC', 'AK', 'BKL', 'DF', 'VASC', 'SCC']

# Chuyển từ One-Hot Encoding sang Single Label
df['label'] = df[classes].idxmax(axis=1)
df['path'] = df['image'].apply(lambda x: os.path.join(DATA_DIR, x + '.jpg'))

print("\nDữ liệu sau khi xử lý:")
display(df[['image', 'label', 'path']].head())

print(f"Tổng số lượng mẫu: {len(df)}")
```

```python
df.drop(['MEL', 'NV', 'BCC', 'AK', 'BKL', 'DF', 'VASC', 'SCC'], axis=1, inplace=True)
```

```python
df
```

```python
# Định nghĩa danh sách các class theo thứ tự mong muốn
classes = ['MEL', 'NV', 'BCC', 'AK', 'BKL', 'DF', 'VASC', 'SCC']

# Tạo mapping từ text sang số
class_to_idx = {cls: idx for idx, cls in enumerate(classes)}
print("Mapping từ nhãn text sang số:")
for cls, idx in class_to_idx.items():
    print(f"  {cls} -> {idx}")

# Tạo cột 'target' bằng cách map cột 'label'
df['target'] = df['label'].map(class_to_idx)

# Kiểm tra kết quả
print("\nDữ liệu sau khi thêm cột target:")
display(df[['image', 'label', 'target', 'path']].head(10))

# Kiểm tra phân bố các class
print("\nPhân bố các class:")
print(df['target'].value_counts().sort_index())
```

```python
df.to_csv('ISIC_2019_test_metadata.csv', index=False)
```
